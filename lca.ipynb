```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import RandomForestClassifier, RandomForestRegressor
from sklearn.feature_selection import SelectFromModel
from sklearn.metrics import accuracy_score, confusion_matrix
from sklearn.model_selection import cross_val_score
```

```python
data = load_breast_cancer()
```

```python
data
```

```python
X = pd.DataFrame(data.data, columns=data.feature_names)
y = pd.Series(data.target)
```

```python
X.describe()
```

```python
X.head()
```

```python
plt.figure(figsize=(10, 6))

sns.kdeplot(X, color='blue')
sns.kdeplot(y, color='red')
```

```python
sns.scatterplot(X)
```

```python
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
plt.figure(figsize=(10, 6))

sns.kdeplot(X_scaled, color='blue')
```

```python
sns.scatterplot(X_scaled)
```

```python
plt.figure(figsize=(20, 10))
sns.heatmap(X.corr(), annot=True, cmap="coolwarm", fmt=".2f")
plt.title('Feature Correlation Heatmap')
plt.show()
```

```python
rf = RandomForestClassifier(n_estimators=100, random_state=42)
rf.fit(X_scaled, y)
```

```python
importances = rf.feature_importances_
indices = np.argsort(importances)[::-1]
```

```python
selector = SelectFromModel(rf, threshold="mean", max_features=10)
X_selected = selector.fit_transform(X_scaled, y)
```

```python
selected_features = X.columns[selector.get_support()]
print("Selected Features:", selected_features)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_selected, y, test_size=0.3, random_state=42)
```

```python
classifier = RandomForestClassifier(n_estimators=100, random_state=42)
classifier.fit(X_train, y_train)
```

```python
y_pred = classifier.predict(X_test)
```

```python
accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy: {accuracy * 100:.2f}%")
```

```python
cm = confusion_matrix(y_test, y_pred)
plt.figure(figsize=(8, 6))
sns.heatmap(cm, annot=True, fmt="d", cmap="Blues", xticklabels=data.target_names, yticklabels=data.target_names)
plt.title("Confusion Matrix")
plt.xlabel("Predicted")
plt.ylabel("True")
plt.show()
```

```python
cv_scores = cross_val_score(classifier, X_selected, y, cv=5)
print(f"CV Scores: {cv_scores}")
print(f"Mean CV Score: {cv_scores.mean():.2f}")
```
